```python
from pathlib import Path
extensions = [".wav", ".mp3"]

def find_audio(age_dir: Path, session_id: str) -> Path | None:
  cand = []
  for p in age_dir.glob(f"{session_id}.*"):
    if p.suffix.lower() in extensions and p.is_file():
      cand.append(p)
    cand = sorted(cand, key=lambda p: (0 if p.suffix.lower()==".wav" else 1, p.suffix.lower()))
    return cand[0] if cand else None
```
